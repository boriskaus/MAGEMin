```python
# Code to generate gradient and hessian matrices from gibbs function using thermocalcal #
# input database. This code uses symbolic python to generate those equations            #

import numpy as np
import math as math
# import scipy as scipy
from scipy.optimize import minimize
from sympy import *

init_printing()


def gem(name,P,T,state):

    #first clean up the end-member file, remove first and last lines of the file
    path = 'tc-ds633_endmembers.txt'
    gem = open(path,'r')

    #declare endmember structure, name + thermodynamic data
    em = {}
    new_content = [[] for x in range(4)] #create 2d array with 4 lines and unspecified column number

    #read thermocalc endmember file and fill endmember structure
    c=1;
    for line in gem:
        words = line.split()
        count = len(words)

        if c == 1: #attribute endmember name to the dictionary
            name0 = words[0]
            content = np.asarray(words[1:count])   
            new_content[0] = [float(i) for i in content]
        elif c > 1 and c < 5: #attribute thermodynamic data to the endmember dictionary entry
            content = np.asarray(words[0:count]) 
            new_content[c-1] = [float(i) for i in content]  

        em[name0] = new_content[0:4]

        c+=1
        if c == 5:
            c = 1

#     check random endmember entries       
#     display(em['fo'])

    # for i in em : 
    #     if i == 'ski':
    #         print(i, em[i][:][:]) 

    #                 THIS HAS TO BE A FUNCTION ARG                   #


    # var('T,P') #here I was trying to have a symbolic formulation for P and T, but the root finder messes it!

#     name = 'fa' #Random endmember here

    t0       = 298.15;
    R        = 0.0083144598;
    enthalpy = em[name][1][0]; entropy  = em[name][1][1]; volume   = em[name][1][2];
    cpa      = em[name][2][0]; cpb      = em[name][2][1]; cpc      = em[name][2][2];
    cpd      = em[name][2][3]; alpha0   = em[name][3][0]; kappa0   = em[name][3][1];
    kappa0p  = em[name][3][2]; kappa0pp = em[name][3][3];

    lgt=len(em[name][:][0]) #size of the endmember composition

    n=0;
    for i in range(2,lgt,2): #find the number of atoms for the endmember
        n = n+em[name][0][i]

    if name == 'L':
        dkappa0dt = em[name][3][4];
        kappat = kappa0 + dkappa0dt*(T - t0);
        gbase = (enthalpy + cpa*(T - t0) + cpb*(T**2 - t0**2)/2 
                - cpc*(1/T - 1/t0) + 2*cpd*(T**(1/2) - t0**(1/2)) 
                - T*(entropy + 2*cpa*(log(T**(1/2)) - log(t0**(1/2))) + cpb*(T - t0) 
                - cpc/2*(T**-2 - t0**-2) - 2*cpd*(T**(-1/2) - t0**(-1/2)))
                + volume*exp(alpha0*(T - t0)) 
                * ((1 + kappa0p*P / kappat)**(1 - 1/kappa0p) - 1)*kappat / (kappa0p - 1));
    else:
        if em[name][3][4] == 0:
            landaut  = 0; smax     = -0.001; vmax     = 0; sfdh     = 0; sfdhv    = 0;
            sfw      = 0; sfwv     = 0; sfn           = 0; sffac    = 0;
        elif em[name][3][4] == 1:
            landaut = em[name][3][5]; smax = em[name][3][6]; vmax = em[name][3][7];
            sfdh     = 0; sfdhv    = 0; sfw      = 0; sfwv     = 0; sfn      = 0;
            sffac    = 0;
        elif em[name][3][4] == 2:
            landaut  = 0; smax                  = -0.001; vmax             = 0;
            sfdh     = em[name][3][5]; sfdhv    = em[name][3][6]; sfw      = em[name][3][7];
            sfwv     = em[name][3][8]; sfn      = em[name][3][9]; sffac    = em[name][3][10];

        #Chemical potential computation 
        theta = round(10636/(entropy*10**3/n + 6.44)); #NOTE: theta should be rounded to nearest integer

        pth   = (theta * alpha0 * kappa0 / (exp(theta/t0) * (theta/t0)**2 / (exp(theta/t0) - 1)**2) 
               *(1/(exp(theta/T) - 1) - 1/(exp(theta/t0) - 1) ));
        ta    = (1 + kappa0p)/(1 + kappa0p + kappa0 * kappa0pp);
        tb    = (kappa0p + kappa0p**2 - kappa0 * kappa0pp)/(kappa0 * (1 + kappa0p));
        tc    = (1 + kappa0p + kappa0 * kappa0pp)/(kappa0p + kappa0p**2 - kappa0 * kappa0pp);

        vterm = P * (1 - ta) + ta * (-(1 + tb * (P - pth))**(1 - tc) + (1 - tb * pth)**(1 - tc))/(tb * (tc - 1));

        gbase = (enthalpy + cpa * (T - t0) + cpb * (T**2 - t0**2)/2 - cpc * (1/T - 1/t0) + 2 
               * cpd * (T**(1/2) - t0**(1/2)) - T * (entropy + 2 * cpa * (log(T**(1/2)) 
               - log(t0**(1/2))) + cpb * (T - t0) - cpc/2 * (T**-2 - t0**-2) - 2 * cpd 
               * (T**(-1/2) - t0**(-1/2))) + volume * vterm);

        if sfn > 0:        
            def func(qq):
                return( (sfdh + P * sfdhv + (sfw + P * sfwv) * (2 * qq - 1) + sffac * sfn/(sfn + 1)
                        * R * T * log(sfn * (1 - qq)**2/((1 + sfn * qq) * (sfn + qq)))) );

            if state == "equilibrium":
                min = 1e-20 #min bound
                max = 1-1e-20 #max bound
                q      = scipy.optimize.brentq(func, min, max);
                sfterm = (sfdh + P * sfdhv + q * (sfw - sfdh + P * (sfwv - sfdhv)) - q**2 * (sfw + P * sfwv) 
                        + T * sffac * R * ((1+sfn * q) * log((1 + sfn * q)/(sfn + 1)) 
                        + sfn * (1 - q) * log((1 - q)/(sfn + 1)) 
                        + sfn * (1 - q) * log(sfn * (1 - q)/(sfn + 1)) 
                        + sfn * (sfn + q) * log((sfn + q)/(sfn + 1))) / (sfn + 1));
                gbase = gbase + sfterm;
            if landaut != 0:

                tc0 = landaut;
                if vmax == 0:  tc  = tc0; 
                else: tc  = tc0 + P * vmax / smax;
                if smax <= 0:  q20 = 0;   
                else: q20 = sqrt(1 - t0 / tc0);
                if T    >  tc: q2  = 0; 
                else: q2  = sqrt((tc - T) / tc0);
                if smax <= 0:  sm  = 0; 
                else: sm  = smax;

                if TF == 'equilibrium':
                    gbase = gbase + sm * ((tc0 * (q20 * (1 - 1/3 * q20^2) + 1/3 * q2^3) - q2 * tc) - T * (q20 - q2)) + p * vmax * q20;
                else:
                    gdis =  sm * ((tc0 * (q20 * (1 - 1/3 * q20^2) + 1/3 * q2^3) - q2 * tc) - T * (q20 - q2)) + P * vmax * q20;
                    a    = 2*smax;
                    gord = 1/2*a*(t-tc)*q2+1/6*a*tc0*q2^3;
                    gbase = gbase + gdis - gord;

#         display(em[name])
        
    #Chemical composition
    el_comp = np.zeros(19);
    chem_comp = np.zeros(11);
    
    y  = 1;
    l = len(em[name][0])-1;

    while y < l:
        el_comp[int(em[name][0][y])-1] = em[name][0][y+1];
        y = y + 2;

    #Normalization of endmember composition
    chem_comp[0] = el_comp[0];
    chem_comp[1] = el_comp[2]/2;
    chem_comp[2] = el_comp[6];
    chem_comp[3] = el_comp[4];
    chem_comp[4] = el_comp[3];
    chem_comp[5] = 0.;
    chem_comp[6] = el_comp[8]/2;
    chem_comp[7] = el_comp[7]/2;
    chem_comp[8] = el_comp[1];
    chem_comp[9] = el_comp[18]/2;
    chem_comp[10] = el_comp[10]/2;

    while (2*chem_comp[0] + 3*chem_comp[1] + chem_comp[2] + chem_comp[3] + chem_comp[4] + 3*chem_comp[5] +  chem_comp[6] + chem_comp[7] + 2*chem_comp[8] + 3*chem_comp[9]+ chem_comp[10]) != el_comp[9]:

        if chem_comp[4] < 0:
            display('stoichiometric error');

        chem_comp[4] = chem_comp[4] - 1;
        chem_comp[5] = chem_comp[5] + 1/2;    

    return( gbase,chem_comp );
```

```python
# open solid solution file
path = 'tc-ax-output_phases.txt'
gss = open(path,'r')

#declare solid-solution structure
info = {};

#read thermocalc solid solution file
c=0;d=1;
for line in gss:
    words = line.split()
    count = len(words)

    if count > 0:
        if words[0] == '--------------------------------------': #This is the (ugly) solid solution separator, do not change the number of dashes or it will get even more ugly
            c = 0;
        else:
            c = c+1;   
            
            if c == 1: #find solid solution name and begining line in the file
                name0 = words[0];
                info[name0] = {};

            if c > 1 and words[0] == 'starting':            
                info[name0]['starting_guess'] = d;
            if c > 1 and words[0] == 'site':            
                info[name0]['site_fractions'] = d;
            if c > 1 and words[0] == 'proportions':            
                info[name0]['proportions'] = d;
            if c > 1 and words[0] == 'ideal':            
                info[name0]['idealmixing'] = d; 
            if c > 1 and words[0] == 'non-ideality':            
                info[name0]['non_ideality'] = d; 
            if c > 1 and words[0] == 'dependent':            
                info[name0]['dependent_em'] = d;             
            if c > 1 and words[0] == 'DQF':            
                info[name0]['dqf_increments'] = d;                         
    
    d=d+1; #Line increment
```

```python
# open solid solution file
path = 'tc-ax-output_phases.txt'
gss = open(path,'r')

ss = {};

#read thermocalc solid solution file
c=0;d=1;

for line in gss:
    words = line.split()
    function = line.strip()
    count = len(words)
    
    if count > 0:
        if words[0] == '--------------------------------------': #This is the (ugly) solid solution separator, do not change the number of dashes or it will get even more ugly
            c = 0;
        else:
            c = c+1;   
            
            if c == 1: #find solid solution name and begining line in the file
                name0 = words[0];
                ss[name0] = {};
                ss[name0]['proportions'] = [];
                ss[name0]['dependent_em'] = [];

            if d > info[name0]['proportions'] and d < info[name0]['idealmixing']-1:
                ss[name0]['proportions'].append(function);
                
            if 'dependent_em' in info[name0]:
                if d > info[name0]['dependent_em'] and d < info[name0]['dqf_increments']-1:
                    ss[name0]['dependent_em'].append(function);        


    d=d+1; #Line increment
    
display(ss)
```

```python
def flt(text):
    lett = '';
    text2 = [];
    text3 = [];

    l = len(text)
    for i, let in enumerate(text):
        lett = let;
        if i > 0 and i < l:
            if text[i-1].isdigit() == true and text[i].isalpha() == true:
                lett = '*'+let;
        if i > 0 and i < (l - 1):
            if text[i-1] != ' ' and text[i] == ' ' and text[i+1] != ' ':
                if text[i-1] != '+' and text[i-1] != '=' and text[i-1] != '-' and text[i-1] != '/':
                    if text[i+1] != '+' and text[i+1] != '=' and text[i+1] != '-' and text[i+1] != '/':
                        lett = ' * ';
        text2.append(lett);
    text3 = (''.join(text2));

    return(text3)
```

```python
symss = {}
for k in ss:
    symss[k] = {};    
    for n in ss[k]:
        symss[k][n] = [];
        for o in ss[k][n]:
                symss[k][n].append(flt(o));       

symss_exp = {}
for k in symss:
    symss_exp[k] = {};    
    for n in symss[k]:
        symss_exp[k][n] = {};
        for o in symss[k][n]:
            tmp = [];
            words = o.split()
            l = len(words); 
            c=0;
            for q in words:
                if c > 1: 
                    tmp.append(q);
                c=c+1;
            rhs = (''.join(tmp));
            symss_exp[k][n][words[0]] = [rhs];    
                
# display(symss_exp)
```

```python
wss  = ['opx','g','spn','cpx','pl','hb']; #wanted set of solid solutions
alpha = len(wss)

rs = []; rms = []; 
tp = []; tmp = [];   

for i in wss:
    for j in symss[i]['dependent_em']:
        tmp = []; 
        words = j.split()
        l = len(words)
        c=0;
        for q in words:
            if c > 1: 
                tmp.append(q);
            c=c+1;
            
        rms = (''.join(tmp));
        raw_emlist = sympify(rms); #turn all the needed endmember as symbols
        str_emlist = raw_emlist.free_symbols #list of symbols (endmembers)
        for q in str_emlist:
            tp.append('+')
            tp.append(str(q))

emd = [];
emp = []; 

for k in wss: 
    for t in symss[k]['dependent_em']:
        words1 = t.split()
        emd.append(words1[0]);
        
    for s in symss[k]['proportions']:   
        words = s.split()
        emp.append(words[0]);

for k in emp:
    if k not in emd:
        tp.append('+');
        tp.append(k);   

rs = (''.join(tp));
raw_emlist = sympify(rs); #turn all the needed endmember as symbols
str_emlist = raw_emlist.free_symbols #list of symbols (endmembers)

emlist = [];
for i in str_emlist:
    emlist.append(str(i)); #create the list of endmembers as string

for i in emlist: #attach to each endmember str a variable and its related composition
    gbase, chem_comp = gem(i,P,T,state); 
    vars()[i] = chem_comp;
```

```python
comp_em = {}
comp_var = {}
for k in wss: 
    if k in symss_exp:
        if not symss_exp[k]['dependent_em']:
            for i in symss_exp[k]['proportions']:
                comp_em[i] = i;
        else:
            for i in symss_exp[k]['proportions']:
                comp_var[i]= symss_exp[k]['proportions'][i];
                if i in symss_exp[k]['dependent_em']:
                    comp_em[i]= symss_exp[k]['dependent_em'][i];
                else:
                    comp_em[i]= [i];

display(comp_em)
display(comp_var)
# display(emlist)
```

```python
# calculate the chemical composition of compositional-variables-related endmembers
fem_comp = {}

for i in comp_em:
    fem_comp[i] = {}
    for j in comp_em[i]:
        fem_comp[i] = eval(j)
# display(fem_comp)
```

```python
#fem_comp, chemical composition  of endmembers
#comp_var, compositional variables function

# G is the equality constraint functions, the RHS here (bulk-rock rock) is not included yet
g = {}
for i in fem_comp:
    g[i] = [];
    g[i] = sympify(comp_var[i])*fem_comp[i];
    
# print(g)
```

```python
t1 = ['0' for x in range(11)]
gcons = sympify(t1)

for p,i in enumerate(g):
    for o,j in enumerate(g[i]):
        if p < 11:
            gcons[o] = g[i][o] + gcons[o];

display(gcons) 
```
